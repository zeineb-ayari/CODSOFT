```python
def choosing1():
    choice=''
    while not(choice.lower()=='rock' or choice.lower()=='paper'or choice.lower()=='scissors'):
        choice=input('enter your choice (rock/paper/scissors): ')
    return choice
```

```python
def choosing2():
    global n
    n=randint(1,3)
    if(n==1):
        return 'rock'
    elif(n==2):
        return 'paper'
    else:
        return 'scissors'
```

```python
def score(m,n):
    global user_score,computer_score
    if(n=='rock'):
        if(m=='rock'):
            user_score,computer_score=1,1
        elif(m=='paper'):
            user_score,computer_score=1,0
        else:
            user_score,computer_score=0,1
    elif(n=='paper'):
        if(m=='rock'):
            user_score,computer_score=0,1
        elif(m=='paper'):
            user_score,computer_score=1,1
        else:
            user_score,computer_score=1,0
    else:
        if(m=='scissors'):
            user_score,computer_score=1,1
        elif(m=='paper'):
            user_score,computer_score=0,1
        else:
            user_score,computer_score=1,0
    
```

```python
from random import*
total_computer_score,total_user_score=0,0
while True:
    print('option 1: continue playing')
    print('option 2: quit playing')
    p=int(input('enter your option: '))
    if(p==1):
        user_choice=choosing1()
        computer_choice=choosing2()
        print('you chose:',user_choice)
        print('the computer chose:',computer_choice)
        score(user_choice,computer_choice)
        total_computer_score+=computer_score
        total_user_score+=user_score
        if(user_score>computer_score):
            print('Congratulations, you win this round')
        elif(user_score<computer_score):
            print('The computer wins this round')
        else:
            print("it's a tie")
    elif(p==2):
        print('you scored',total_user_score)
        print('the computer scored',total_computer_score)
        if(total_user_score>total_computer_score):
            print('Congratulations, you win the overall game')
        elif(total_user_score<total_computer_score):
            print('The computer wins the overall game')
        else:
            print("it's a tie in the overall game")
        break
    else:
        print('invalid choice')
```
